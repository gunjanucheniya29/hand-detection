```python
import cv2
import numpy as np
import matplotlib.pyplot as plt






```

```python
Image = cv2.imread("cv2.jpg")
```

```python
img_rgb = cv2.cvtColor(Image,cv2.COLOR_BGR2RGB)
plt.Figure(figsize=(6,4))
plt.imshow(img_rgb)
plt.title("1.Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(Image,cv2.COLOR_BGR2GRAY)
plt.Figure(figsize=(6,4))
plt.imshow(gray, cmap ='gray')
plt.title("2. Grayscale Image")
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
Image = cv2.imread("cv2.jpg")
mg_shape = Image.copy()

img_shape = Image.copy()
cv2.rectangle(img_shape,(50,50),(150,150),(0,255,0),3)
cv2.circle(img_shape,(200,200),40,(255,0,0),-1)
cv2.putText(img_shape,"Hello",(10,400),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
shape_rgb = cv2.cvtColor(img_shape,cv2.COLOR_BGR2RGB)
plt.Figure(figsize=(6,4))
plt.imshow(shape_rgb)
plt.title("5. Shape and Text")
plt.axis("off")
plt.show()
```
